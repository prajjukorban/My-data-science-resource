## multipal plots in one

```python
fig,((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2)
ax1.plot(["a","b"],[5,6])
ax1.set_title("line plot")
ax2.bar(["a","b"],[5,6])
ax2.set_title("bar chart")
ax3.scatter(["a","b","c","d"],[5,6,3,2.5])
ax3.set_title("scatter plot")
ax4.hist([5,6,7],bins=5);
ax4.set_title("histogram")
```

## try another dataset

```python
heart_data = pd.read_csv("heart-disease.csv")
```

```python
heart_data.head()
```

```python
#create a hist of age
heart_data["age"].plot.hist(bins=100)
```

```python
heart_data.plot.hist(figsize=(10,50),subplots=True);
```

```python
over50 = heart_data[heart_data["age"]>50]
```

```python
over50.head()
```

```python
over50.plot(kind="scatter",x="age",y="chol",c="target")
```

```python
fig,ax = plt.subplots()
 
scatter = ax.scatter(x=over50["age"],
           y=over50["chol"],
           c=over50["target"],
                    cmap="winter")

ax.legend(*scatter.legend_elements())

ax.axhline(over50["chol"].mean(),linestyle="--");
```
